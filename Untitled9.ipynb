## Analyze distributions

### Subtask:
Explore the distributions of key features using visualizations.


**Reasoning**:
Create histograms for selected numerical features and bar plots for selected categorical features to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_features = ['AGE', 'BODY_WEIGHT', 'TOTAL_LENGTH_OF_STAY', 'COST_OF_IMPLANT']
categorical_features = ['GENDER', 'MARITAL_STATUS', 'KEY_COMPLAINTS_CODE', 'IMPLANT_USED_']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_features):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Data preprocessing

### Subtask:
Handle missing values and encode categorical features.


**Reasoning**:
Impute missing numerical and categorical values and then handle categorical features by either one-hot encoding or dropping based on the number of unique values as instructed.



```python
for col in numerical_features:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

for col in categorical_features:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

categorical_cols_to_encode = []
categorical_cols_to_drop = []

for col in categorical_features:
    if df[col].nunique() <= 10:
        categorical_cols_to_encode.append(col)
    else:
        categorical_cols_to_drop.append(col)

df = pd.get_dummies(df, columns=categorical_cols_to_encode, drop_first=True)
df = df.drop(columns=categorical_cols_to_drop)

display(df.head())
display(df.isnull().sum())
```

**Reasoning**:
Some numerical columns still have missing values after the previous imputation step. Impute these remaining missing numerical values with the mean of their respective columns, and then check for any remaining missing values.



```python
numerical_cols_with_missing = ['BP_HIGH', 'BP_LOW', 'HB', 'UREA', 'CREATININE']

for col in numerical_cols_with_missing:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

display(df.isnull().sum())
```

**Reasoning**:
The categorical columns with more than 10 unique values were not dropped in the previous step. Drop the 'PAST_MEDICAL_HISTORY_CODE' column as it has more than 10 unique values, and then confirm that all missing values have been handled and the categorical features have been processed as instructed.



```python
df = df.drop(columns=['PAST_MEDICAL_HISTORY_CODE'])
display(df.isnull().sum())
display(df.head())
```

## Model evaluation

### Subtask:
Evaluate the performance of the model using appropriate metrics.


**Reasoning**:
Calculate and print the evaluation metrics (MAE, MSE, R-squared) for the linear regression model using the test set.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2 ): {r2:.2f}')
```
